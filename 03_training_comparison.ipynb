# Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Configuration

```python
SCENARIO = 1

"""
1 : TF-IDF + Naive Bayes
2 : TF-IDF + SVM
3 : FastText + SVM
4 : IndoBERT
"""
```

```python
SCENARIO_CONFIG = {

    1: {
        "name": "TF-IDF + Naive Bayes",
        "file": "results/traning/tfidf_nb_results.csv"
    },

    2: {
        "name": "TF-IDF + SVM",
        "file": "results/traning/tfidf_svm_results.csv"
    },

    3: {
        "name": "FastText + SVM",
        "file": "results/traning/fasttext_svm_results.csv"
    },

    4: {
        "name": "IndoBERT",
        "file": "results/traning/indobert_results.csv"
    }

}

MODEL_NAME = SCENARIO_CONFIG[SCENARIO]["name"]
RESULT_FILE = SCENARIO_CONFIG[SCENARIO]["file"]

print(MODEL_NAME)
```

```python
results = pd.read_csv(
    RESULT_FILE
)

results
```

# Visual Result

## Accuracy

```python
plt.figure(figsize=(7,5))

plt.plot(
    results["Train_Size"],
    results["Accuracy"],
    marker="o"
)

plt.xlabel("Training Data Size")
plt.ylabel("Accuracy")

plt.title(
    f"Accuracy - {MODEL_NAME}"
)

plt.grid()

plt.show()
```

## Precision

```python
plt.figure(figsize=(7,5))

plt.plot(
    results["Train_Size"],
    results["Precision"],
    marker="o"
)

plt.xlabel("Training Data Size")
plt.ylabel("Precision")

plt.title(
    f"Precision - {MODEL_NAME}"
)

plt.grid()

plt.show()
```

## Recall

```python
plt.figure(figsize=(7,5))

plt.plot(
    results["Train_Size"],
    results["Recall"],
    marker="o"
)

plt.xlabel("Training Data Size")
plt.ylabel("Recall")

plt.title(
    f"Recall - {MODEL_NAME}"
)

plt.grid()

plt.show()
```

## F1

```python
plt.figure(figsize=(7,5))

plt.plot(
    results["Train_Size"],
    results["F1-Score"],
    marker="o"
)

plt.xlabel("Training Data Size")
plt.ylabel("F1 Score")

plt.title(
    f"F1 Score - {MODEL_NAME}"
)

plt.grid()

plt.show()
```

# Best Model

```python
best = results.loc[
    results["F1-Score"].idxmax()
]

print("=== BEST RESULT ===\n")

print(best)
```

# Save Result

```python
os.makedirs("results/comparisson", exist_ok=True)

OUTPUT = f"results/comparisson/scenario_{SCENARIO}.csv"

results.to_csv(
    OUTPUT,
    index=False
)

print("Saved:", OUTPUT)
```
