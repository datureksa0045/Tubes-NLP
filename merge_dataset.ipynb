```python
import os
import pandas as pd

# 1. Daftar semua nama file hasil terjemahan dari anggota kelompok
# Sesuaikan nama file di bawah ini dengan nama file asli kalian
files = [
    "chunk_1_indonesia.csv",
    "chunk_2_indonesia.csv",
    "chunk_3_indonesia.csv",
    "chunk_4_indonesia.csv",
    "chunk_5_indonesia.csv",
    "chunk_6_indonesia.csv",
]

# 2. Baca semua file dan kumpulkan ke dalam sebuah list
df_list = []
for file in files:
    if os.path.exists(file):
        df_part = pd.read_csv(file)
        df_list.append(df_part)
        print(f"Berhasil membaca {file} dengan jumlah baris: {len(df_part)}")
    else:
        print(f"Peringatan: File {file} tidak ditemukan!")

# 3. Gabungkan semua data frame menjadi satu
if df_list:
    df_gabungan = pd.concat(df_list, ignore_index=True)

    # 4. Simpan hasil akhir ke file baru
    nama_file_akhir = "Dataset_News_Full_Indonesia.csv"
    df_gabungan.to_csv(nama_file_akhir, index=False, encoding="utf-8-sig")

    print("\n==========================================")
    print(f"Selesai! Semua data telah digabungkan.")
    print(f"Total baris setelah digabung: {len(df_gabungan)}")
    print(f"File disimpan dengan nama: {nama_file_akhir}")
    print("==========================================")
    print(df_gabungan.head())
else:
    print("Tidak ada data yang berhasil digabungkan.")
```
