```python
import pandas as pd

# 1. Baca file CSV yang ingin diacak
# Ganti 'dataset_input.csv' dengan nama file CSV kamu
df = pd.read_csv('golden_dataset/golden_dataset_Full.csv')

# 2. Acak dataset
# frac=1 artinya kita mengambil 100% data untuk diacak
# random_state=42 (opsional) fungsinya agar hasil acakan konsisten setiap kali kode dijalankan
df_acak = df.sample(frac=1, random_state=42).reset_index(drop=True)

# 3. Simpan hasil acakan ke file CSV baru
df_acak.to_csv('golden_dataset/random_dataset.csv', index=False)

print("Dataset berhasil diacak dan disimpan sebagai 'random_dataset.csv'!")
```
