# Chunk dataset Indo

```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"dataset\dataset.csv", sep=",", engine="python")

# Ukuran dataset bertingkat
chunk_sizes = [10000, 20000, 30000, 40000]

for size in chunk_sizes:
    # Ambil data dari awal hingga ukuran tertentu
    chunk = df.iloc[:min(size, len(df))]

    # Simpan file
    filename = f"new_chunk_translate_{size//1000}k.csv"
    chunk.to_csv(filename, index=False, encoding="utf-8-sig")

    print(f"{filename} -> {len(chunk)} data")
```
