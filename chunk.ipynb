# Chunk dataset Indo

```python
## chunk data 2
import pandas as pd
import numpy as np

df = pd.read_csv(r"dataset\dataset.csv", sep=",", engine="python")
chunk_sizes = [10000]
chunks = []
start_idx = 0
i = 1

for size in chunk_sizes:
    end_idx = start_idx + size
    if end_idx <= len(df):
        chunks.append(df.iloc[start_idx:end_idx])
        start_idx = end_idx
    else:
        # Jika sisa data kurang dari ukuran yang diinginkan, ambil sisa yang ada
        chunks.append(df.iloc[start_idx:])
        break

chunks.to_csv(f"new_chunk_translate_{i}.csv", index=False, encoding="utf-8-sig")
print(f"new_chunk_translate_{i}.csv -> {len(chunks)} data")
```
