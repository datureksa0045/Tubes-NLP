```python
import pandas as pd
import numpy as np

df = pd.read_csv("News_Title_Indonesia_Rename.csv")

# Acak dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

chunks = np.array_split(df, 6)

for i, chunk in enumerate(chunks, start=1):
    chunk.to_csv(
        f"chunk_{i}.csv",
        index=False,
        encoding="utf-8-sig"
    )

    print(f"chunk_{i}.csv -> {len(chunk)} data")
```
